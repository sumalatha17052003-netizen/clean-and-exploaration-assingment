```python
df["transmission_type"]=df["transmission_type"].astype(str)
df["transmission_type"].dtype
df["transmission_type"]=df["transmission_type"].str.strip().str.title()
df["transmission_type"]=df["transmission_type"].replace({"Automatic": 1, "Manual": 0})
print(df["transmission_type"].head(20))


```

```python
a=pd.get_dummies(df["fuel_type"],drop_first=True)
b=pd.concat([df,a],axis=1)
b.drop("fuel_type",axis=1,inplace=True)
b.head()

```

```python
from sklearn.model_selection import train_test_split
x=df.drop("selling_price", axis=1)
y=df["selling_price"]
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)

```

```python
from sklearn.metrics import mean_absolute_error
baseline_pred=np.mean(y_train)
y_pred_baseline=np.repeat(baseline_pred,len(y_test))
baseline_mae=mean_absolute_error(y_test,y_pred_baseline)
print(f"Baseline MAE: ₹{round(baseline_mae)}")  
```
